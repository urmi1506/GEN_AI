```python
def remove_html_tags(text):
    pattern=re.compile('<.*?>')
    return pattern.sub("",text)
```

```python
def remove_url(text):
    pattern=re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub("",text)
```

```python
exclude=string.punctuation
```

```python
for char in exclude:
    print(char)
```

```python
def remove_punc(text):
    for char in exclude:
        text=text.replace(char,"")
    return text
```

```python
remove_punc(text)
```

```python
text2="my name is s@nny sa##vita!!!!!!!!!"
```

```python
remove_punc(text2)
```
