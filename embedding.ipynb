1) Different texts → different embeddings (bcz model encode meaning not word)
   Same model → same embedding size

2) why embedding comparision req ---measure semantic closeness(how twwo thing related to each other)
   --measure high / low similarity  (high == most relevant , low =less relevant)
   what happen whn we not compare ?
   -semantic search fail  in search s/y (search something ..but not found most relevant ans)
   -RAG fail (as its working is Question → Embedding → Compare → Retrieve → LLM → Answer)
      --no relevant context found by LLM 
      --hallucinations increase ,give wrong ans
   -Recommendation s/y fail (not give relevant recommendation)


```python
# import libraries
from sentence_transformers import SentenceTransformer ,util

# load model
model =SentenceTransformer("all-MiniLM-L6-v2")

# input text 
text1 = "AI is useful"
text2 ="AI helps developers"
text3 = "I like mangoes"

# generate embedding
embedding1 =model.encode(text1)
embedding2 =model.encode(text2)
embedding3 =model.encode(text3)

print("embedding1 length :",len(embedding1))
print("embedding2 length :",len(embedding2))

print("Text1 first 3 values:", embedding1[:3])
print("Text2 first 3 values:", embedding2[:3])

# compare similarity
sim_1_2 = util.cos_sim(embedding1 ,embedding2)
sim_1_3 = util.cos_sim(embedding1 ,embedding3)


print("Similarity b/w text1 & 2 :",sim_1_2.item())
print("Similarity b/w text1 & 3 :",sim_1_3.item())
```
